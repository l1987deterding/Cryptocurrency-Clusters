```python
# import dependencies
import pandas as pd
```

```python
# read in csv file
crypto = pd.read_csv('crypto_data.csv')
```

```python
# observe dataframe
crypto
```

```python
# discard all cryptocurrencies that are not being traded
crypto_clean = crypto[crypto['IsTrading'] == True]
crypto_clean
```

```python
crypto_clean = crypto_clean.drop(columns='IsTrading')
```

```python
crypto_clean
```

```python
crypto_clean = crypto_clean.dropna(inplace = False)
```

```python
crypto_clean
```

```python
# Filter for cryptocurrencies that have been mined. That is, the total coins mined should be greater than zero.
crypto_cleaned=crypto_clean['TotalCoinsMined']>0
crypto_cleaned = crypto_clean[crypto_cleaned]
crypto_cleaned
```

```python
# delete the CoinName from the original dataframe.
crypto_cleaned = crypto_cleaned.drop(columns='CoinName')
crypto_cleaned
```

```python
# Convert Algorithm column data to numerical data
crypto_cleaned = pd.get_dummies(crypto_cleaned, columns = ['Algorithm'])
```

```python
# Convert ProofType data into numberical data
crypto_cleaned = pd.get_dummies(crypto_cleaned, columns = ['ProofType'])
```

```python
crypto_cleaned
```

There are now 99 columns compared to 5 columns before. This is a HUGE difference! 

```python
# delete Unnamed: 0 from the dataframe.
crypto_cleaned = crypto_cleaned.drop(columns='Unnamed: 0')
crypto_cleaned
```

```python
# Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(crypto_cleaned)
crypto_cleaned_scaled = scaler.transform(crypto_cleaned)
crypto_cleaned_scaled
```

```python
# Perform PCA to reduce dimensions; keep 90% of explained variance
from sklearn.decomposition import PCA
pca = PCA(n_components=0.90)
```

```python
crypto_pca = pca.fit_transform(crypto_cleaned_scaled)
```

```python
### BILL: Make sure I did this correctly
# Transform PCA data to a DataFrame
crypto_pca = pd.DataFrame(
    data=crypto_pca)
crypto_pca.head()
```

Now there are 74 columns instead of 99

```python
# Run t-SNE
from sklearn.manifold import TSNE
tsne = TSNE(learning_rate=200)
tsne_features = tsne.fit_transform(crypto_pca)
```

```python
# Check that dimensions have been reduced to two
tsne_features.shape
```

```python
# Plot the results
import matplotlib.pyplot as plt
plt.scatter(tsne_features[:,0], tsne_features[:,1])
plt.show()
```

```python
# Cluster analysis with k-means
from sklearn.cluster import KMeans
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_pca)
    inertia.append(km.inertia_)

# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plot the elbow curve to find the best candidate(s) for k
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow curve for customer data')
plt.show()
```

Recommendation: There's no one group size better than the other. There are not distinct clusters, so you can choose the number of groups that you want according to what you need.
